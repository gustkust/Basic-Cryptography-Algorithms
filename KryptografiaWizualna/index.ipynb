```python
from matplotlib import pyplot as plt
from skimage import img_as_float
import skimage.io as io
from random import randint


img = img_as_float(io.imread('test.png'))

# biały (1)
C0 = [
    [1, 0],
    [0, 1]
]

# czarny (0)
C1 = [
    [
        [1, 0],
        [0, 1]
    ],
    [
        [0, 1],
        [1, 0]
    ]
]


new_img1 = []
new_img2 = []

for i in img:
    row_img1 = []
    row_img2 = []
    for j in i:
        if j[0] == 1:
            if randint(0, 1) == 1:
                row_img1.append(1)
                row_img1.append(0)
                row_img2.append(1)
                row_img2.append(0)
            else:
                row_img1.append(0)
                row_img1.append(1)
                row_img2.append(0)
                row_img2.append(1)
        elif j[0] == 0:
            if randint(0, 1) == 1:
                row_img1.append(0)
                row_img1.append(1)
                row_img2.append(1)
                row_img2.append(0)
            else:
                row_img1.append(1)
                row_img1.append(0)
                row_img2.append(0)
                row_img2.append(1)
    new_img1.append(row_img1)
    new_img2.append(row_img2)

res = []
i = 0
while i < len(new_img1):
    row = []
    j = 0
    while j < len(new_img1[i]):
        if new_img1[i][j] + new_img2[i][j] >= 1:
            row.append(1)
        else:
            row.append(0)
        j += 1
    res.append(row)
    i += 1
        

fig, axs = plt.subplots(1, 4)
axs[0].imshow(img, cmap="gray")
axs[1].imshow(new_img1, cmap="gray")
axs[2].imshow(new_img2, cmap="gray")
axs[3].imshow(res, cmap="gray")


fig.set_size_inches(14, 10)
plt.show()
```
